```python
import shutil

def zip_folder(folder_path, output_path):
    """
    Compresses the folder at folder_path into a zip file saved at output_path.
    """
    shutil.make_archive(output_path, 'zip', folder_path)

# Example usage
folder_to_zip = '/kaggle/working/UCMCTrack'
output_zip = '/kaggle/working/UCMCTrack'
zip_folder(folder_to_zip, output_zip)
```
